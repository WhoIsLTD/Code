```python
import numpy as np
import matplotlib.pyplot as plt
st = np.loadtxt('E:/Code/Python/Tài liệu Python/KhoaHocDuLieu/Slide_Baigiang/Data_C4/thongke.txt')
x = ['Hà Nội', "Vinh", 'Đà Nẵng', "Nha Trang", "Hồ Chí Minh", "Cà Mau"]
y = st[0,:6]
plt.subplots(figsize = (15, 8))
plt.grid(True)
plt.plot(x, y, 'r*-.', linewidth = 2,
         markersize = 15, label = "ToC - Max")
plt.ylabel("Nhiệt độ", size = 18)
plt.xticks(size = 15)
plt.yticks(size = 15)
plt.legend()
plt.title("BIỂU ĐỒ BIỂU DIỄN NHIỆT ĐỘ MAX CỦA 6 THÀNH PHỐ", size = 20)
plt.show()
```

```python
y = st[1,:6]
plt.subplots(figsize = (15, 8))
plt.bar(x, y, color = 'black')
plt.grid(True)
plt.ylabel("Nhiệt độ", size = 18)
plt.xticks(size = 15)
plt.yticks(size = 15)
plt.title("BIỂU ĐỒ BIỂU DIỄN NHIỆT ĐỘ MEAN CỦA 6 THÀNH PHỐ", size = 20)
plt.show()
```

```python
y = st[2,:6]
plt.subplots(figsize = (15, 8))
plt.scatter(x, y, marker = 'o', color = 'blue', s = 400, alpha = 0.25)
plt.grid(True)
plt.ylabel("Nhiệt độ", size = 18)
plt.xticks(size = 15)
plt.yticks(size = 15)
plt.title("BIỂU ĐỒ BIỂU DIỄN NHIỆT ĐỘ MIN CỦA 6 THÀNH PHỐ", size = 20)
plt.show()
```
