#### Yêu cầu 3:

```python
#Thay thế các giá trị missing ở cột ‘loai’ bằng giá trị cố định là một chuỗi: ‘Asymptomatic’
df['loai'] = df['loai'].fillna('Asymptomatic')
df.iloc[292:293,:]
```

```python
#Thay thế các giá trị missing ở cột ‘thalassemia’ bằng số 3
df['thalassemia'] = df['thalassemia'].fillna(3)
df.iloc[3:4,:]
```

```python
df['gioitinh'] = df['gioitinh'].map({'Male' : 0, 'Female' : 1}).astype(int)
df
```

```python
df.groupby('loai').count()
```

#### Yêu cầu 4

```python
convert = {'Atypical angina' : 0, 'Typical angina' : 1, 'Asymptomatic' : 2, 'Non-anginal pain' : 3}
df['loai'] = df['loai'].map(convert).astype(int)
df
```
